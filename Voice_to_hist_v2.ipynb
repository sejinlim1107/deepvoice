```python
import numpy as np
import scipy.io as sio
import scipy.io.wavfile
import matplotlib.pyplot as plt
import os

def extract_features(data_path, save_path, filenames):
    for fn in filenames:
      rate, data = sio.wavfile.read(data_path + '/' + fn) # reading wave file.
      fn = fn.rstrip('.wav')
      plt.hist(data, bins='auto')
      plt.savefig(save_path + '/' + fn + '.png')
      plt.show()
```
